```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
X = np.random.normal(size=N)
```

```python
def F(x):
    return np.sin(x) + np.log(np.maximum(x, 0.1))

def L(y, x, noise_std):
    return (y - F(x))**2 / noise_std**2
```

```python
noise_std = 1e-1
Y = F(X) + np.random.normal(size=X.shape[0], scale=noise_std)
```

```python
x = np.linspace(-10, 10, 1000)
y = -1
ell = L(y, x, noise_std)
plt.plot(x, ell)
# plt.yscale("log")
```
